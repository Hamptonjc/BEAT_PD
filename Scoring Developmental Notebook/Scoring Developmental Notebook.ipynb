```python
# Create data directory path
datadir = 'Data/BEATPDchallenge/'

# load cis-pd labels
TestLabelsREALDf = pd.read_csv(datadir + 'REAL-PD_Test_Data_IDs.csv')
display(TestLabelsREALDf.head())

# load cis-pd labels
TestLabelsCISDf = pd.read_csv(datadir + 'CIS-PD_Test_Data_IDs.csv')
TestLabelsCISDf.head()
```

```python
PredDf = pd.concat([TestLabelsCISDf,TestLabelsREALDf], 
                   ignore_index=True)
isCIS = np.zeros(len(PredDf), dtype = int )
isCIS[:len(TestLabelsCISDf)] = 1
PredDf['CIS'] = isCIS
PredDf
```

```python
## Manufactured Data to test Scoring

actuals = []
for i,row in PredDf.iterrows():
    if( row.CIS ):
        actuals.append(np.random.randint(0,5))
    else:
        actuals.append(np.random.randint(0,2))
actuals = np.array( actuals, dtype = float)
PredDf['actual'] = actuals

preds = []
for i,row in PredDf.iterrows():
    if( row.CIS ):
        preds.append(np.random.randint(0,5))
    else:
        preds.append(np.random.randint(0,2))
preds = np.array(preds, dtype = float)

# α = confidence of exact

α = 0.85

PredDf['predicted'] = α*actuals + (1-α)*preds
PredDf.head()
```
